## 予測モデル1

### 決定木(Decision Tree)---不純度が最も減少するように条件分岐を作りデータを振り分ける教師あり機械学習手法です。
#### 決定木とは木構造を用いて分類や回帰を行う機械学習の手法の一つです
#### 分類木と回帰木の総称して決定木といいます．
#### 名前の通り分類木は対象を分類する問題を解き，回帰木は対象の数値を推定する問題を解きます．

```python
from sklearn import tree
```

「Pclass」「Sex」「Age」「Fare」を使用して生存を予測（「チケットクラス（社会経済的地位）」「性別」「年齢」「料金」）

#### 「train」の目的変数(出力)と説明変数(入力)の値を取得

```python
target = train["Survived"].values
#target
```

```python
features_one = train[["Pclass", "Sex", "Age", "Fare"]].values
features_one
```

#### 決定木の作成
DecisionTreeClassifier()

```python
tree_one = tree.DecisionTreeClassifier()
#DecisionTreeClassifier（）を使ってtree_oneという決定木モデルを作成
type(tree_one)
```

#### 学習するデータを追加
fit()

```python
tree_one = tree_one.fit(features_one, target)
```

#### 予測したいデータを追加して予測
predict()

```python
test_features = test[["Pclass", "Sex", "Age", "Fare"]].values
```

```python
my_prediction = tree_one.predict(test_features)
```

```python
my_prediction.shape
```

```python
my_prediction
```

```python
test_prediction_Survived=test
test_prediction_Survived['Survived']=my_prediction
test_prediction_Survived
```

```python
test_prediction_Survived.to_csv('test_tree_one.csv')
```
